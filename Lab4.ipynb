```python
import numpy as np 
import pandas as pd

f = lambda x: np.sin(x)
g = lambda x: np.sin(1000*x)

df = lambda x: np.cos(x)
dg = lambda x: 1000*np.cos(1000*x)

enumf=[]
enumg=[]
eexaf = []
eexag=[]

aproxf = [0]
aproxg = [0]


enumcf=[]
enumcg=[]

eexacf=[]
eexacg=[]

aproxcf=[0]
aproxcg=[0]


for i in range(1, 17):
    h = 10**(-i)
    a = 1/2
    diff =  (f(a+h) - f(a))/h 
    difg =  (g(a+h) - g(a))/h 
    difcf = (f(a+h)-f(a-h))/(2*h)
    difcg = (g(a+h)-g(a-h))/(2*h)
    
    aproxf.append(diff)
    aproxg.append(difg)
    aproxcf.append(difcf)
    aproxcg.append(difcg)
    
    errnumf = abs((diff-aproxf[i-1])/diff)
    errnumg = abs((difg-aproxg[i-1])/difg)
    
    errnumcf = abs((difcf-aproxcf[i-1])/difcf)
    errnumcg = abs((difcg-aproxcg[i-1])/difcg)
    
    errexaf = abs((df(a)-diff)/df(a))
    errexag = abs((dg(a)-difg)/dg(a))
    
    errexacf = abs((df(a)-difcf)/df(a))
    errexacg = abs((dg(a)-difcg)/dg(a))
    
    enumf.append(errnumf)
    enumg.append(errnumg)
    eexaf.append(errexaf)
    eexag.append(errexag)
    
    enumcf.append(errnumcf)
    enumcg.append(errnumcg)
    eexacf.append(errexacf)
    eexacg.append(errexacg)
```

```python
dff = pd.DataFrame()
dfg = pd.DataFrame()
dfcf = pd.DataFrame()
dfcg = pd.DataFrame()
```

```python
dff["Error exacto f"]=(eexaf)
dff["Error numérico f"]=(enumf)
dfg["Error exacto g"]=(eexag)
dfg["Error numérico g"]=(enumg)

dfcf["Error exacto f centrada"]=(eexacf)
dfcf["Error numérico f centrada"]=(enumcf)
dfcg["Error exacto g centrada"]=(eexacg)
dfcg["Error numérico g centrada"]=(enumcg)
```
